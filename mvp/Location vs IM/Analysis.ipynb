```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
infant_state_df = pd.read_csv("Infant_state.csv")
infant_state_df.head()
```

```python
pure_data = infant_state_df.drop(columns=['URL'])
```

```python
year_data = pure_data[:50]
```

```python
year_data
```

```python
data_sorted = year_data.sort_values('RATE', ascending= False)
```

```python
data_sorted
```

```python
bottom_three = data_sorted.nsmallest(3, 'RATE')
```

```python
bottom_three
```

```python
top_three = data_sorted.nlargest(3, 'RATE')
```

```python
top_three
```

```python
x_axis = np.arange(len(top_three))
tick_location = [value+0.4 for value in x_axis]
```

```python
fig = plt.figure(figsize=(20,3))

plt.bar(x_axis, top_three["RATE"], color='r', alpha=0.5, align = "center")
plt.bar(x_axis, bottom_three["RATE"], color='b', alpha=0.9, align = "center")
plt.xticks(tick_locations, top_three["STATE"], rotation="horizontal")
plt.ylabel('Rate per 1000')
plt.xlabel('State')
plt.title('Top 3 States')
```

```python
x_axis = np.arange(len(data_sorted))
tick_locations = [value+0.4 for value in x_axis]
```

```python
fig = plt.figure(figsize=(20,3))

plt.bar(x_axis, data_sorted["RATE"], color='r', alpha=0.5, align="edge")
plt.xticks(tick_locations, data_sorted["STATE"], rotation="horizontal")
plt.ylabel('Rate per 1000')
plt.xlabel('State')
plt.title('Infant Mortality by State')
```
